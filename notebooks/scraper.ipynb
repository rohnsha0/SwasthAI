```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
appollo_webpage=requests.get('https://www.askapollo.com/physical-appointment/city/kolkata?speciality=General%20Physician&page=1').text
```

```python
appollo_webpage
```

```python
appollo_soup= BeautifulSoup(appollo_webpage,'lxml')
```

```python
doctor_details= appollo_soup.find_all('div', class_='dr-list')
len(doctor_details)
```

```python
doctor_names= []
for i in doctor_details:
    doctor_names.append(i.find_all('h3')[0].text.strip())
```

```python
doctor_names
```

```python
spec_grp= appollo_soup.find_all('div', class_='spec-group')
```

```python
doctor_spec= []
for i in spec_grp:
    doctor_spec.append(i.text.strip().split('|'))
```
