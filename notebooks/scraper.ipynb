```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_doctor_data(page_url):
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    doctor_data = []
    doctor_cards = soup.find_all('div', class_='listing-doctor-card')
    
    for card in doctor_cards:
        name = checkNullity(card.find('h2', class_='doctor-name'))
        specialty = checkNullity(card.find('div', class_='u-grey_3-text').find('div', 'u-d-flex'))
        experience = checkNullity(card.find('div', {'data-qa-id': 'doctor_experience'}))
        location = checkNullity(card.find('span', {'data-qa-id': 'practice_locality'}))
        fee = checkNullity(card.find('span', {'data-qa-id': 'consultation_fee'}))
        doctor_data.append([name, specialty, experience, location, fee])
    
    return doctor_data

def checkNullity(text):
    if text is None:
        return text
    return text.text.strip()
```

```python
base_url = "https://www.practo.com/kolkata/doctors?page="
all_doctors = []
for page_num in range(1, 100):  # Assuming there are 34 pages
    print("-------------------------")
    print(f"starting pg: {page_num}")
    page_url = base_url + str(page_num)
    doctors = get_doctor_data(page_url)
    all_doctors.extend(doctors)
    print(f"ended pg: {page_num}")
    print("-------------------------")
    
print(len(all_doctors))
```
